# # 跨行业股票财务指标对比分析（静态数据集版）
# 
# **目的**：比较科技、金融、消费行业头部公司的估值水平和股东回报。
# **数据来源**：从 Yahoo Finance 收集的快照数据（保存为 CSV，避免 API 限流）
# **分析日期**：2026-04-21

```python
# %% [code]
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 1. 读取 CSV 数据（假设文件在同一目录下）
df = pd.read_csv('financial_data.csv')
print("原始数据：")
print(df)

# 2. 数据清洗：将 PEG 列的缺失值填充为 0（或保留，但分析时不影响）
df_clean = df.copy()
df_clean['PEG'] = df_clean['PEG'].fillna(0)
print("\n清洗后数据（PEG 缺失已填充）：")
print(df_clean)

# 3. 可视化1：市盈率对比
plt.figure(figsize=(10,6))
sns.barplot(data=df_clean, x='股票代码', y='市盈率(PE)', hue='行业', palette='Set2')
plt.title('各公司市盈率对比', fontsize=14)
plt.ylabel('市盈率')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 4. 可视化2：市净率对比
plt.figure(figsize=(10,6))
sns.barplot(data=df_clean, x='股票代码', y='市净率(PB)', hue='行业', palette='Set3')
plt.title('各公司市净率对比', fontsize=14)
plt.ylabel('市净率')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 5. 可视化3：股息率对比
plt.figure(figsize=(10,6))
sns.barplot(data=df_clean, x='股票代码', y='股息率(%)', hue='行业', palette='pastel')
plt.title('各公司股息率对比', fontsize=14)
plt.ylabel('股息率 (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 6. 散点图：市盈率 vs 股息率（气泡大小=市净率）
plt.figure(figsize=(10,6))
scatter = plt.scatter(df_clean['市盈率(PE)'], df_clean['股息率(%)'], 
                      s=df_clean['市净率(PB)']*10, c=range(len(df_clean)), cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='市净率(映射)')
for i, row in df_clean.iterrows():
    plt.annotate(row['股票代码'], (row['市盈率(PE)'], row['股息率(%)']), fontsize=9)
plt.xlabel('市盈率')
plt.ylabel('股息率 (%)')
plt.title('市盈率 vs 股息率 (气泡大小=市净率)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 7. 输出结论表格
print("\n===== 分析结论 =====")
print(df_clean[['股票代码', '行业', '市盈率(PE)', '市净率(PB)', '股息率(%)']])

# 8. 保存清洗后的数据（可选）
df_clean.to_csv('stock_comparison_cleaned.csv', index=False)
print("\n清洗后数据已保存为 stock_comparison_cleaned.csv")
```
